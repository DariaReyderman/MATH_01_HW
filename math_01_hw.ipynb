```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
def linear_function1(x):
    # x + y = 4
    return -1 * x + 4

def linear_function2(x):
    # 2x - y = 2
    return 2 * x - 2

'''
x + y = 4
2x - y = 2
'''

x_values = np.linspace(-100, 200, 50)
y_values1 = linear_function1(x_values)
y_values2 = linear_function2(x_values)
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")  # y-axis
plt.plot(x_values, y_values1, color="green")
plt.plot(x_values, y_values2, color="blue")

plt.title("Exercise 1")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-10, 20)
plt.ylim(-20, 20)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

x_intersection = 2
y_intersection = 2
plt.scatter(x_intersection, y_intersection, color="red", s=100, edgecolors="black", label="Intersection")
plt.text(x_intersection + 1, y_intersection, f"({x_intersection}, {y_intersection})", fontsize=12, color="red")


plt.show()
```

```python
def linear_function1(x):
    # 2x + y = 6
    return 6 - 2 * x

def linear_function2(x):
    # 4x + 2y = 10
    return -1 * x + 5

'''
2x + y = 6
4x + 2y = 10
'''

x_values = np.linspace(-100, 200, 50)
y_values1 = linear_function1(x_values)
y_values2 = linear_function2(x_values)
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")  # y-axis
plt.plot(x_values, y_values1, color="yellow")
plt.plot(x_values, y_values2, color="orange")

plt.title("Exercise 2")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

x_intersection = 1
y_intersection = 4
plt.scatter(x_intersection, y_intersection, color="red", s=100, edgecolors="black", label="Intersection")
plt.text(x_intersection + 1, y_intersection, f"({x_intersection}, {y_intersection})", fontsize=12, color="red")


plt.show()
```
